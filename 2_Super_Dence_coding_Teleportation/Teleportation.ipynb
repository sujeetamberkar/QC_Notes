```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
circuit=QuantumCircuit(3,3)
circuit.x(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1,2) ### To entangle
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit=QuantumCircuit(3,3)
circuit.x(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1,2) ### To entangle
circuit.barrier()
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit=QuantumCircuit(3,3)
circuit.x(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1,2) ### To entangle
circuit.barrier()
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
#### Now Alice will measure the qubits and put in the clasical bits
circuit.measure([0,1],[0,1])
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit=QuantumCircuit(3,3)
circuit.x(0)
circuit.barrier()
circuit.h(1)
circuit.cx(1,2) ### To entangle
circuit.barrier()
circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
#### Now Alice will measure the qubits and put in the clasical bits
circuit.measure([0,1],[0,1])
circuit.barrier()
# Now change q2 according to q0 and q1
circuit.cx(1,2)
circuit.cz(0,2)
circuit.barrier()
circuit.measure([2],[2])
circuit.draw(output='mpl')
```

```python
Aer.backends()
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=simulator,shots=1024).result()
counts=result.get_counts(circuit)
plot_histogram([counts])
#plot_histogram(result.get_counts(circuit))
```
